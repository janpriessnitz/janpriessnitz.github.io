```python
from bs4 import BeautifulSoup
import base64
```

```python
with open('show.html') as html_fp:
  soup = BeautifulSoup(html_fp, 'html.parser')
```

```python
plot_elems = soup.find_all(attrs={'class': 'prizz_plot'})
```

```python
p = plot_elems[0]
```

```python
ext_fname = p.attrs['src']
```

```python
with open(ext_fname, 'rb') as ext_fp:
  ext_bytes = ext_fp.read()
```

```python
ext_bytes
```

```python
b = base64.b64encode(ext_bytes)
b
```

```python
s = b.decode()
```

```python
new_attr = "data:text/plain;base64,"+s
```

```python
p.attrs['src'] = new_attr
```

```python
with open("output1.html", "w", encoding='utf-8') as file:
    file.write(str(soup))
```
